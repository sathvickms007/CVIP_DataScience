```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset - netflix1.csv')
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
sb.countplot(x='type', data=df)
plt.title('count of movies & TV shows on netflix')
plt.xticks(rotation=45)
plt.show()
```

```python
df.shape
```

```python
rat_counts = df['rating'].value_counts()

plt.pie(rat_counts, labels=rat_counts.index, autopct='%1.1f%%')
plt.title('Piechart of ratings on netflix')
plt.show()
```

```python
df['release_year'] = df['release_year'].astype(str)
cby = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sb.lineplot(x=cby.index, y=cby.values)
plt.title('Netflix content released over the years')
plt.xticks(rotation=45)
plt.show()
```

```python
sb.set(style="darkgrid")
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(10, 6))  
sb.barplot(x=rating_counts.index, y=rating_counts.values, palette="viridis")  # Create the bar plot
plt.title('Distribution of content ratings on netflix') 
plt.xlabel('Content rating')
plt.ylabel('Count') 
plt.xticks(rotation=45) 
plt.show()
```

```python
df.info()
```

```python
df['release_year'] = df['release_year'].astype(int)
plt.figure(figsize=(10, 6))
sb.violinplot(y='release_year', data=df[df['type'] == 'Movie'])
plt.title('Distribution of movie release years on netflix')
plt.ylabel('relese years')
plt.show()
```
